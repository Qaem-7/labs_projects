```python
# Creating time categories function to return Morning, Noon, Evenging, Night based on transaction_time
def assign_time_category(hour):
    if 4 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Noon'
    elif 17 <= hour < 20:
        return 'Evening'
    else:
        return 'Night'



# use apply to apply the time category into a new column called time_category
coffee_df['time_category'] = coffee_df['hour'].apply(assign_time_category)
```

##### 3.❓Do customers who have been customers longer tend to spend more? Plot a chart to show the result (Hint : You may need to use date.today() to get todays date)

```python
# Do customers who have been customers longer tend to spend more?
loyal_df = coffee_df.dropna(subset=['customer_since']).copy()
loyal_df['days_as_customer'] = (pd.to_datetime('2026-07-08') - loyal_df['customer_since']).dt.days

customer_spent = loyal_df.groupby('days_as_customer')['line_item_amount'].sum().reset_index()
customer_spent.plot(kind='scatter', x='days_as_customer',y='line_item_amount')
plt.show()
```

### Findings:
Inspecting the scatter plot above, it can be seen that customers loyalty does not correlate with there spending on the coffee shop. The data widespread with either new or old customers, this indicates that customers only strive to achieve their needs rather than loyalty towards the coffee shop.
