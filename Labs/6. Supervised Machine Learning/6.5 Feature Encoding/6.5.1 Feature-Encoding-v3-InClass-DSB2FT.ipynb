## 2. One-Hot Encoder

A **One-Hot Encoder** is a data preprocessing tool used to convert **categorical data into a numerical format**. It transforms text labels or categories into binary vectors (columns of 0s and 1s) so that mathematical algorithms can interpret them correctly.

---

### How One-Hot Encoding Works
The encoder creates a **separate binary column for every unique category** in the original feature. 
* A **1** represents the presence of a category.
* A **0** represents its absence.

#### Example: Encoding a "Fruit" Feature
If you have a column named `Fruit` with three distinct categories (`Apple`, `Mango`, and `Orange`), a One-Hot Encoder expands that single column into three separate columns:

| Original Column (`Fruit`) | New Column (`Fruit_Apple`) | New Column (`Fruit_Mango`) | New Column (`Fruit_Orange`) |
| :--- | :---: | :---: | :---: |
| Apple | **1** | 0 | 0 |
| Mango | 0 | **1** | 0 |
| Orange | 0 | 0 | **1** |

## Nominal vs. Ordinal Categories

1. **Nominal Categories**: 
    - Nominal categories are unordered categories that represent distinct groups or labels without any inherent order or ranking. 
    - Each category is treated as a separate and independent entity. 
    - Examples of nominal categories include colors (e.g., red, blue, green), countries (e.g., USA, Canada, Australia), or types of animals (e.g., cat, dog, bird). 
    - Nominal categories can be encoded using techniques like `one-hot encoding` or `get_dummies`, where each category is represented by a binary indicator variable. 



2. **Ordinal Categories**:
    - Ordinal categories, on the other hand, carry information about the relative order or ranking of the categories. 
    - They represent categories with a meaningful order or hierarchy. 
    - Each category has an assigned value or rank that signifies its position in the order. 
    - Examples of ordinal categories include ratings (e.g., low, medium, high), education levels (e.g., elementary, high school, college), or clothing sizes (e.g., small, medium, large). 
    - Ordinal categories can be encoded using techniques like `ordinal encoding`, where each category is mapped to a numerical value based on its order or rank.



> It's important to understand whether your categorical data represents nominal or ordinal information, as it influences the appropriate encoding and analysis techniques to be used.
